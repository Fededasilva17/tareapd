```python
import pandas as pd
import plotly.express as px

df=pd.read_csv("cars.csv")

# Realizamos el melt para organizar los datos
df_melted = df.melt(id_vars=['safety'], 
                    value_vars=["doors", "lug_boot", "persons"],
                    var_name='Feature', 
                    value_name='Value')

# Creamos el gr√°fico de barras apiladas
fig = px.bar(df_melted, x='Feature', y='Value', color='safety',
             title='Distribuci√≥n de Caracter√≠sticas seg√∫n Seguridad',
             labels={'Feature': 'Caracter√≠stica', 'Value': 'Valor'},
             barmode='stack')

fig.show()
```
